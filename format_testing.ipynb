```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
```

```python
# Formatting 
background_vagabond_colors = ["#fffdfa", "#fef8f0", "#fbe4cb", "#e8c29d", "#9a7b56", "#2b322e"]
vagabond_colors = ["#ae9e98", "#9f8f7f", "#884c45", "#5a2028",
                    "#924a42", "#a4b9b1", "#4ea7a6", "#026d73", "#06494f"]

rc('font', **{'family':'serif', 'serif':['Computer Modern']})
rc('text', usetex = True)
rc('axes', edgecolor = 'black')
```

```python
plt.rcParams['axes.facecolor'] = background_vagabond_colors[0]
x = np.linspace(0, 10, 100); 

fig, ax = plt.subplots(layout='constrained')

for i in range(0, len(vagabond_colors), 1):
    y = 0 * x + i; 
    ax.plot(x, y, color = vagabond_colors[i], linewidth = 3); 

ax.set_xlabel('x axis', fontsize = 15); 
ax.set_ylabel('y axis', fontsize = 15);  
ax.set_title("title", fontsize = 20); 

ax.tick_params(axis='both', width = 1, size = 4, labelsize = 12); 

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(1)  
```
